```
good_proportions_gear = []
 
for t_s in range(min_sun_teeth, max_sun_teeth + 1):
  for t_p in range(min_planet_teeth, max_planet_teeth + 1):
    for t_r in range(min_ring_teeth, max_ring_teeth + 1):
      if t_s+2*t_p-t_r == 0:
        good_proportions_gear.append([t_s, t_p, t_r])
```

```
def gearRatio(first, secound):
  try:
    return (1+first[2]/secound[2])/(1-(secound[1]*first[2])/(secound[2]*first[1]))  
  except:
    pass

good_ratios_gear = []
 
for first_gear in good_proportions_gear:
  for secound_gear in good_proportions_gear:
      ratio = gearRatio(first_gear, secound_gear)
      if ratio == None:
        pass
      elif ratio in ratios:
        good_ratios_gear.append([first_gear, secound_gear])

good_ratios_gear
```

```
def gearTwoModule(first, secound, module):
  return (first[0]+first[1])/(secound[0]+secound[1])*module

for good_gear in good_ratios_gear:
  print(gearTwoModule(good_gear[0], good_gear[1], 1))
```
