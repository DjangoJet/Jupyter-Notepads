```python
import numpy as np
import random
```

```python
class Game:
  def __init__(self, max_x, max_y):
    self.max_x = max_x - 1
    self.max_y = max_y - 1
    self.board = np.zeros((max_x, max_y))
    self.game_end = False
    self.last_board = None

  def randNewLocation(self):
    self.last_board = self.board
    temp_flag = True
    while(temp_flag):
      x = random.randint(0,self.max_x)
      y = random.randint(0,self.max_y)
      if(self.board[x,y] == 0):
        self.board[x, y] = random.randint(1,2)*2
        temp_flag = False

  def checkIfFull(self):
    if 0 in self.board:
      return False
    else:
      return True

  def undo(self):
    if(self.last_board != None):
      self.board = self.last_board
      self.last_board = None

  def shift(self, direction):
    if(direction == "right"):
      self.shiftRight()
    elif(direction == "left"):
      self.shiftLeft()
    elif(direction == "up"):
      self.shiftUp()
    elif(direction == "down"):
      self.shiftDown()
    self.randNewLocation()
    

  def shiftRight(self):
    new_board = np.zeros_like(self.board)
    for x in range(self.max_x + 1):
      last_pos = self.max_y
      for y in range(self.max_y, -1, -1):
        if(self.board[x,y] != 0):
          if(new_board[x, last_pos] == self.board[x,y]):
            new_board[x,last_pos] = new_board[x,last_pos] + self.board[x,y]
          elif(new_board[x, last_pos] == 0):
            new_board[x,last_pos] = self.board[x,y]
          else:
            last_pos = last_pos - 1
            new_board[x,last_pos] = self.board[x,y]
    if(~(np.allclose(self.board, new_board))):
      self.last_board = self.board
      self.board = new_board
          
  def shiftLeft(self):
    new_board = np.zeros_like(self.board)
    for x in range(self.max_x + 1):
      last_pos = 0
      for y in range(self.max_y + 1):
        if(self.board[x,y] != 0):
          if(new_board[x, last_pos] == self.board[x,y]):
            new_board[x,last_pos] = new_board[x,last_pos] + self.board[x,y]
          elif(new_board[x, last_pos] == 0):
            new_board[x,last_pos] = self.board[x,y]
          else:
            last_pos = last_pos + 1
            new_board[x,last_pos] = self.board[x,y]
    if(~(np.allclose(self.board, new_board))):
      self.last_board = self.board
      self.board = new_board

  def shiftUp(self):
    new_board = np.zeros_like(self.board)
    for y in range(self.max_y + 1):
      last_pos = 0
      for x in range(self.max_y + 1):
        if(self.board[x,y] != 0):
          if(new_board[last_pos,y] == self.board[x,y]):
            new_board[last_pos,y] = new_board[last_pos,y] + self.board[x,y]
          elif(new_board[last_pos,y] == 0):
            new_board[last_pos,y] = self.board[x,y]
          else:
            last_pos = last_pos + 1
            new_board[last_pos,y] = self.board[x,y]
    if(~(np.allclose(self.board, new_board))):
      self.last_board = self.board
      self.board = new_board

  def shiftDown(self):
    new_board = np.zeros_like(self.board)
    for y in range(self.max_y + 1):
      last_pos = self.max_x
      for x in range(self.max_x, -1, -1):
        if(self.board[x,y] != 0):
          if(new_board[last_pos,y] == self.board[x,y]):
            new_board[last_pos,y] = new_board[last_pos,y] + self.board[x,y]
          elif(new_board[last_pos,y] == 0):
            new_board[last_pos,y] = self.board[x,y]
          else:
            last_pos = last_pos - 1
            new_board[last_pos,y] = self.board[x,y]
    if(~(np.allclose(self.board, new_board))):
      self.last_board = self.board
      self.board = new_board
```
