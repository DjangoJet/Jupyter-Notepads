```python
import numpy as np
from keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(train_X, train_y), (test_X, test_y) = mnist.load_data()
```

```python
print(train_X.shape)
print(train_y.shape)
print(test_X.shape)
print(test_y.shape)
```

```python
def binarizedImage(img, treshold):
  temp_img = np.zeros_like(img)
  for i in range(0, img.shape[0]):
    for j in range(0, img.shape[1]):
      if img[i,j] >= treshold:
        temp_img[i,j] = 1
      else:
        temp_img[i,j] = 0
  return temp_img
```

```python
plt.imshow(train_X[0,:,:], cmap='gray')
```

```python
def binarizedDataset(dataset, treshold):
  temp_train_X_bin = np.zeros_like(dataset)
  for i in range(0, dataset.shape[0]):
    temp_train_X_bin[i,:,:] = binarizedImage(dataset[i,:,:], treshold)
  return temp_train_X_bin
```

```python
train_X_bin = binarizedDataset(train_X, 64)
```

```python
plt.imshow(train_X_bin[0], cmap="gray")
```

```python
def createFrequencyArray(dataset_bin, labels, number):
  label_count = 0
  temp_frequency_array = np.zeros_like(dataset_bin[0,:,:])
  for index, label in enumerate(labels):
    if label == number:
      label_count += 1
      for i in range(0, dataset_bin[index,:,:].shape[0]):
        for j in range(0, dataset_bin[index,:,:].shape[1]):
          if dataset_bin[index,i,j] == 1:
            temp_frequency_array[i,j] += 1
  temp_frequency_array = temp_frequency_array / label_count
  return temp_frequency_array
```

```python
frequency_array_1 = createFrequencyArray(train_X_bin, train_y, 1)
```

```python
test_max = np.amax(frequency_array_1)
factor = 255 / test_max
test = frequency_array_1 * factor
test = test.astype("uint8")
plt.imshow(test, cmap="gray")
```
